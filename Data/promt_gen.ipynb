```python
import pandas as pd
from underthesea import sent_tokenize
import time
def summarize_text(text, temperature=1, instructions=None):
    try:
        # delay 1 second between each request
        time.sleep(3)
        if instructions is None: 
            instructions = f"Viết một đoạn văn nêu lên những thông tin quan trọng về {text}."
        else:
            instructions = instructions + text
        model_config = {
            "temperature": temperature,
            # "max_output_tokens": max_tokens
        }
        # BLOCK_NONE for no safety settings
        model = genai.GenerativeModel("gemini-1.5-flash", generation_config=model_config)
        response = model.generate_content(instructions, safety_settings="BLOCK_NONE")
        print(response.text)
        return response.text
    except Exception as e:
        print(f"Error occurred: {e}")
        print(f"Text: {text}")
        print(f"Temperature: {temperature}")
        print(response)
        return None

def make_dataset(texts, instructions=None):
    dataset = pd.DataFrame(columns=['title', 'gemini_text', 'gpt3_sentences', 'annotation', 'gemini_text_samples'])
    rows = []
    for text in texts:
        gemini_text = summarize_text(text, temperature=0, instructions=instructions)
        # Using underthesea to tokenize text into sentences
        gemini_sentences = sent_tokenize(gemini_text)
        # set annotation to null for each sentence
        annotation = [None] * len(gemini_sentences)
        gemini_text_samples = []
        for i in range(20):
            gemini_text_samples.append(summarize_text(text, temperature=1, instructions=instructions))

        row = {
            'title': text,
            'gemini_text': gemini_text,
            'gemini_sentences': gemini_sentences,
            'annotation': annotation,
            'gemini_text_samples': gemini_text_samples
        }
        rows.append(row)
    dataset = pd.DataFrame(rows)
    return dataset

        

```
